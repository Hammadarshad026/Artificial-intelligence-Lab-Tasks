# Data Pre-Processing

```python
df.info()
```

```python
def fillNaObject(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['street','city','statezip','country']
fillNaObject(columns)
```

```python
def fillNaInt(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['bedrooms','bathrooms','floors','waterfront','view','yr_built']
fillNaInt(columns)
```

```python
def fillNaFloat(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['price','sqft_living','sqft_lot','sqft_above','sqft_basement']
fillNaFloat(columns)
```

```python
df.info()
```

```python
def convertFloattoInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['price','bedrooms','bathrooms','floors','waterfront','view','yr_built','sqft_living','sqft_lot','sqft_above','sqft_basement']
convertFloattoInt(columns)
```

```python
df.info()
```

```python
df.drop('street', axis=1, inplace=True)
```

```python
df.drop('country', axis=1, inplace=True)
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df_encoded = df.copy()
```

```python
def convertObjtoInt(cols):
    for i in cols:
        datalabelEncoder = LabelEncoder()
        df_encoded[i] = datalabelEncoder.fit_transform(df[i])

columns = ['city','statezip']
convertObjtoInt(columns)
```

# Train-Test Split

```python
trainData, testData = train_test_split(df_encoded, test_size=0.2, shuffle=False)
```

```python
# train_x = df_encoded.drop('price')
train_x = trainData.iloc[:,1:]
train_x.head(2)
```

```python
train_y = trainData.iloc[:, 0]
train_y.head(2)
```

```python
test_x = testData.iloc[:,1:]
test_x.head(2)
```

```python
test_y = testData.iloc[:, 0]
test_y.head(2)
```

# Model Training using SVC

```python
# # Initialize the Linear Regression model
regressor = LinearRegression()

# Train the model on the training data
regressor.fit(train_x, train_y)
```

```python
# print("-- Training using SVC on Training Data --")
# print("-- Parameters & Values: ", end='')

# # model_svc = SVC(gamma='auto', random_state=0)
# # model_svc.fit(train_x, np.ravel(train_y))

# model_svc = SVC()
# model_svc.fit(train_x, train_y)

# print(model_svc)
```

```python
# Saving Trained Model
pickle.dump(regressor, open('regressor.pkl', 'wb'))
```

```python
# Load saved Model
model = pickle.load(open('regressor.pkl', 'rb'))
```

```python
model_predictions = regressor.predict(test_x)
```

```python
# RÂ² Score
r2 = r2_score(test_y, model_predictions)

# Mean Squared Error
mse = mean_squared_error(test_y, model_predictions)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# Mean Absolute Error
mae = mean_absolute_error(test_y, model_predictions)

print(f"RÂ² Score: {r2}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Absolute Error: {mae}")
```
